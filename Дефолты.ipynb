```python
import pandas as pd

#Пункт 1

df = pd.read_csv('defaults.csv', sep=',')
df = df.set_index('ID')
print('Размер таблицы: ' + f'{df.shape}')
print()
display(df.iloc[0:10])
display(df.sample(n=7))
```

```python
#Пункт 2

print('Кол-во пропущенных значений (значения нет в ячейке): ' + f'{df.isna().sum().sum()}')

lst = ['EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
counter = 0
for i in range(len(lst)):
    k = df[df[[lst[i]]] == 0][lst[i]].value_counts()[0]
    counter += k
print('Всего нулей в тех столбцах, в которых они по условию задачи не должны быть: ' + f'{counter}')
```

```python
#Пункт 6

import matplotlib.pyplot as plt
import numpy as np

no_default = df['default.payment.next.month'].value_counts()[0]
yes_default = df['default.payment.next.month'].value_counts()[1]
print('Кол-во клиентов без дефолта (0): ' + f'{no_default}')
print('Кол-во клиентов с дефолтом (1): ' + f'{yes_default}')

names = ['С дефолтом', 'Без дефолта']
numbers = np.array([yes_default, no_default])

percent = numbers/(num)*100

new_labels = [i+'  {:.2f}%'.format(j) for i, j in zip(names, percent)]

plt.figure(figsize = (12,3))
plt.barh(new_labels, numbers, color='lightgreen')
plt.title('Распределение клиентов по дефолту')
plt.xlabel('Кол-во клиентов')
```

```python
#Пункт 7

sum_no_def = df[df['default.payment.next.month'] == 0]['LIMIT_BAL']
sum_yes_def = df[df['default.payment.next.month'] == 1]['LIMIT_BAL']

plt.figure(figsize = (15,3))
plt.hist([sum_no_def, sum_yes_def], bins = 90, label=['без дефолта', 'с дефолтом'])
plt.legend(loc='upper right')
plt.title('Суммы кредитов клиентов по категории "Дефолты"')
plt.xlabel('сумма кредита')
plt.ylabel('кол-во клиентов')
plt.show()
```

```python
#Пункт 8.1

mdma = sum_yes_def.mode()[0]
print('Чаще всего происходят дефолты по кредитам размера: ' + f'{mdma}')

#Пункт 8.2

nope = 0
yespe = 0
for elem in sorted(list(set(df['LIMIT_BAL']))):
    nope += int(pd.DataFrame(sum_no_def).value_counts()[elem])
    yespe += int(pd.DataFrame(sum_yes_def).value_counts()[elem])
    if nope > yespe:
        coca = elem
        break
print()
print('На основе графика - начиная с суммы ' + f'{coca}' + ' клиентов без дефолта становится больше, чем клиентов с дефолтом.')

#Пункт 8.3

x = sorted(list(set(sum_no_def.to_list())))
y = sorted(list(set(sum_yes_def.to_list())))
alw_no_def = []
alw_yes_def = []
for i in range(len(x)):
    if x[i] not in y:
        alw_no_def.append(x[i])
for j in range(len(y)):
    if y[j] not in x:
        alw_yes_def.append(y[j])
print()
print('Суммы, по которым никогда не бывает дефолтов:')
print(', '.join(map(str, alw_no_def)))
print()
print('Суммы, по которым всегда дефолт:')
print(', '.join(map(str, alw_yes_def)))
```

```python
#Пункт 9

import matplotlib.pyplot as plt
import seaborn as sns
df_corr = df[['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']].corr()
sns.heatmap(df_corr, cmap='YlGnBu', annot=True, lw=2)
plt.title('Корреляционная матрица')
print('На матрице виден общий тренд по по убыванию корреляции по времени (если смотреть от апреля к сентябрю).')
```

```python
#Пункт 10

from collections import defaultdict
def function(df):
    ansdct = defaultdict(lambda: defaultdict(lambda: 0))
    for elem in sorted(list(set(df['LIMIT_BAL']))):
        num_clients = df['LIMIT_BAL'].value_counts()[elem]
        if elem not in alw_no_def and elem not in alw_yes_def:
            clients_without_def = int(pd.DataFrame(sum_no_def).value_counts()[elem])
            clients_with_def = int(pd.DataFrame(sum_yes_def).value_counts()[elem])
            ratio = clients_with_def/clients_without_def
        elif elem in alw_no_def:
            ratio = 0
        elif elem in alw_yes_def:
            ratio = 1
        ansdct[elem]['ratio'] = ratio
        ansdct[elem]['num_clients'] = num_clients
        sorteddct = sorted(ansdct.items(), key=lambda x: x[1]['ratio'], reverse=True)
    return sorteddct
#перед тем, как использовать функцию, пожалуйста, скомпилируйте все предыдущие пункты.

print('Если не считать ту сумму, по которой всегда дефолты, то есть 327680.0 (1 клиент), то наибольшая доля клиентов с дефолтами у суммы кредита 740000.')
print('В датасете клиентов с такой суммой всего: 2')
print('Вывод о том, что такие суммы кредитов лучше не предоставлять - неверный, так как выборка всего из 2 человек нерепрезентативна.')
```
